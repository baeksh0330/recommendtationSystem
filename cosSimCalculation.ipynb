```python
import pandas as pd
import numpy as np

games = pd.read_csv('C:/MachineLearning TP/recommendtationSystem/game dataset/games_preprocessing.csv')
```

```python
games = games.reset_index(drop=False)
games.head()
```

```python
title = games['title']
games2 = games.drop('title', axis=1)
item_sim = games2.dot(games2.T) / np.linalg.norm(games2, axis=1)
```

```python
def cosSim_get_similar_games(game_id, item_sim, games):
  target_row = games.loc[games["app_id"] == game_id]

  target_row_index = target_row.index[0]

  target_row_sim = item_sim[target_row_index]

  top_3_indices = target_row_sim.argsort()[-4:][::-1]

  recGames = games.iloc[top_3_indices]
  recGames = recGames[recGames['app_id']!=game_id]
  return  recGames



# preferred game id
game_id = 578080
games[games['app_id']==game_id]
```
